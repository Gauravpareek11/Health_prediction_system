```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Health_test/data.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(labels=['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
df.describe()
```

```python
df.shape
```

```python
df['diagnosis'].value_counts() #M=cancerous
```

```python
sns.countplot(df['diagnosis'])
```

```python
dic={'M':1,'B':0}
for i in range(len(df)):
  df['diagnosis'][i]=dic[df['diagnosis'][i]]
```

```python
df['diagnosis']=df['diagnosis'].astype('int64')
```

```python
df.info()
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,15))
matrix = np.triu(df.corr())
sns.heatmap(df.corr(),mask=matrix,annot=True)
```

```python
ax=df.hist(figsize=(20,15))
```

```python
mean_col = ['diagnosis','radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']

sns.pairplot(df[mean_col],hue = 'diagnosis', palette='rocket')
```

```python
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

# Logistic Regression

```python
log=LogisticRegression()
log.fit(x_train,y_train)
log.score(x_train,y_train)
```

```python
accuracy_score(y_test,log.predict(x_test))
```

```python
confusion_matrix(y_test,log.predict(x_test))
```

```python
print(classification_report(y_test,log.predict(x_test)))
```

# kNN

```python
kn=KNeighborsClassifier()
kn.fit(x_train,y_train)
kn.score(x_train,y_train)
```

```python
accuracy_score(y_test,kn.predict(x_test))
```

```python
confusion_matrix(y_test,kn.predict(x_test))
```

```python
print(classification_report(y_test,kn.predict(x_test)))
```

# RandomForest

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
accuracy_score(y_test,rf.predict(x_test))
```

```python
confusion_matrix(y_test,rf.predict(x_test))
```

```python
print(classification_report(y_test,rf.predict(x_test)))
```

# Decision Tree

```python
tr=DecisionTreeClassifier()
tr.fit(x_train,y_train)
tr.score(x_train,y_train)
```

```python
accuracy_score(y_test,tr.predict(x_test))
```

```python
confusion_matrix(y_test,tr.predict(x_test))
```

```python
print(classification_report(y_test,tr.predict(x_test)))
```

# SVC

```python
sv=SVC()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)
```

```python
accuracy_score(y_test,sv.predict(x_test))
```

```python
confusion_matrix(y_test,sv.predict(x_test))
```

```python
print((classification_report(y_test,sv.predict(x_test))))
```

# Naive Bayes

```python
nb=GaussianNB()
nb.fit(x_train,y_train)
nb.score(x_train,y_train)
```

```python
accuracy_score(y_test,nb.predict(x_test))
```

```python
confusion_matrix(y_test,nb.predict(x_test))
```

```python
print(classification_report(y_test,nb.predict(x_test)))
```
